```python
#Text extraction from PDF
def extract_text_from_pdf(pdf_path):
    """
    Extracts text from a PDF resume.
    Args:
        pdf_path (str): Path to resume PDF file
    Returns:
        str: Complete extracted text
    """
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
```

```python
#skill extraction using regex
SKILLS_DB = [
    "python", "machine learning", "deep learning",
    "sql", "numpy", "pandas", "tensorflow",
    "pytorch", "scikit-learn", "nlp"
]

def extract_skills(text):
    """
    Extracts technical skills using regex matching
    """
    text = text.lower()
    skills_found = []

    for skill in SKILLS_DB:
        pattern = rf"\b{skill}\b"
        if re.search(pattern, text):
            skills_found.append(skill)

    return list(set(skills_found))
```

```python
#Named Entity Recognition (AI/NLP)
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    """
    Extracts entities using spaCy NER
    """
    doc = nlp(text)
    entities = {
        "PERSON": [],
        "ORG": [],
        "GPE": []
    }

    for ent in doc.ents:
        if ent.label_ in entities:
            entities[ent.label_].append(ent.text)

    return entities
```

```python
PDF_PATH = "Inderjeet_kaur_resume.pdf"

resume_text = extract_text_from_pdf(PDF_PATH)
skills = extract_skills(resume_text)
entities = extract_entities(resume_text)

print("Extracted Skills:")
print(skills)

print("\nNamed Entities:")
print(entities)
```
